Task 1: Iris Dataset Analysis

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load Iris dataset
df = sns.load_dataset('iris')
```

```python
# Display dataset structure
print("Shape of dataset:", df.shape)
print("\nColumns in dataset:", df.columns)
print("\nFirst 5 rows:\n", df.head())
```

```python
# Scatter plot - petal_length vs petal_width
sns.scatterplot(data=df, x='petal_length', y='petal_width', hue='species')
plt.title("Petal Length vs Width by Species")
plt.show()
```

```python
# Histogram - sepal_length
plt.hist(df['sepal_length'], bins=15, color='skyblue', edgecolor='black')
plt.title("Histogram of Sepal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Frequency")
plt.show()
```

```python
# Box plot - Sepal Width by Species
sns.boxplot(x='species', y='sepal_width', data=df)
plt.title("Boxplot of Sepal Width by Species")
plt.show()
```
